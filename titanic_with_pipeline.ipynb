```python
import pandas as pd 
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectKBest,chi2
```

```python
df=pd.read_csv('D:/train.csv')
df_test=pd.read_csv('D:/test.csv')
```

```python
df.head(5)
```

```python
x_train,y_train=df[['Pclass','Sex','Age','SibSp','Parch','Embarked','Fare']],df[['Survived']]
```

```python
trf1=ColumnTransformer([('impute_age',SimpleImputer(),[2]),('impute_em',SimpleImputer(strategy='most_frequent'),[5])],remainder='passthrough')
```

```python
trf2=ColumnTransformer([('ohe_sex_em',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1,5])],remainder='passthrough')
```

```python
trf3=ColumnTransformer([('scale',MinMaxScaler(),slice(0,9))])
```

```python
trf4=SelectKBest(score_func=chi2,k=7)
```

```python
trf5=RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
pipe=Pipeline([('trf1',trf1),('trf2',trf2),('trf3',trf3),('trf4',trf4),('trf5',trf5)])
```

```python
from sklearn import set_config
set_config(display='diagram')
```

```python
pipe.fit(x_train,y_train)
```

```python
x_test=df_test[['Pclass','Sex','Age','SibSp','Parch','Embarked','Fare']]
```

```python
y_pred=pipe.predict(x_test)
```

```python
y_pred
```

```python
result=df_test[['PassengerId']]
result['Survived']=pd.DataFrame(y_pred)
```
