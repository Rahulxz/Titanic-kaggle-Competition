```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
```

```python
df_train=pd.read_csv('D:/train.csv')
df_test=pd.read_csv('D:/test.csv')
```

```python
df_train.head()
```

```python
df_train=df_train.drop(columns=['Name','PassengerId','Ticket','Fare','Cabin'])
```

```python
df_train
```

```python
df_train.isnull().sum()
```

```python
df_train=df_train.dropna(subset=['Embarked'])
```

```python
x_train,y_train=df_train.drop(columns=['Survived']),df_train[['Survived']]
```

```python
df_test_=df_test.drop(columns=['Name','PassengerId','Ticket','Cabin','Fare'])
```

```python
df_test_.isnull().sum()
```

```python
si_age=SimpleImputer()
x_train_age=si_age.fit_transform(x_train[['Age']])
```

```python
oh_sex=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
oh_em=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
x_train_sex=oh_sex.fit_transform(x_train[['Sex']])
x_train_embarked=oh_em.fit_transform(x_train[['Embarked']])
```

```python
x_rem=x_train[['Pclass','SibSp','Parch']]
x_transformed=np.concatenate((x_rem,x_train_age,x_train_sex,x_train_embarked),axis=1)
x_transformed
```

```python
m_m=MinMaxScaler()
x_train_ready=m_m.fit_transform(x_transformed)
```

```python
clf=DecisionTreeClassifier()
clf.fit(x_train_ready,y_train)
```

```python
x_different_test=df_test_[['Pclass','SibSp','Parch']]
x_diff_sex=oh_sex.transform(df_test[['Sex']])
x_diff_age=si_age.transform(df_test[['Age']])
x_diff_em=oh_em.transform(df_test[['Embarked']])
x_final_test=np.concatenate((x_different_test,x_diff_sex,x_diff_age,x_diff_em),axis=1)
y_test_pred=clf.predict(x_final_test)
```

```python
df_test
```

```python
result=df_test[['PassengerId']]
result['Survived']=pd.DataFrame(y_test_pred)
result
```

```python
result.to_csv('Predicted.csv',index=False)
```
